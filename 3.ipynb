```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = pd.Series(wine.target)
```

```python
df = X.copy()
df['target'] = y
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='alcohol', y='malic_acid', hue='target', data=df, palette='deep')
plt.title('Alcohol vs Malic Acid (Colored by Class)')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['alcohol'].hist(bins=20, color='skyblue')
plt.title('Histogram of Alcohol Content')
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(14,12))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Wine Features')
plt.show()
```

```python
sns.pairplot(df[['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'target']], hue='target', palette='Set2')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='target', y='color_intensity', data=df, palette='pastel')
plt.title('Boxplot of Color Intensity across Different Wine Classes')
plt.xlabel('Wine Class')
plt.ylabel('Color Intensity')
plt.show()
```
